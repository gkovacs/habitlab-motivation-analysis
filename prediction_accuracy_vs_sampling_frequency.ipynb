```python
user_to_difficulty_item_pairs = {}
for user in users_with_difficulty:
  difficulty_items = get_choose_difficulty_items_for_user(user)
  difficulty_items = [x for x in difficulty_items if x.get('is_random') == False]
  if len(difficulty_items) > 0:
    difficulty_item_pairs = [(x['timestamp'],x['difficulty']) for x in difficulty_items]
    user_to_difficulty_item_pairs[user] = difficulty_item_pairs
```

```python
set([y for x,y in difficulty_item_pairs])
```

```python
num_minutes = 43200 # a month = 60*24*30
minutes_to_percent_correct = [0]*num_minutes
for minutes in range(0, num_minutes):
  if minutes % 100 == 0:
    print(minutes)
  interval = 1000*60*minutes # one hour
  num_correct = 0
  num_incorrect = 0
  for user,difficulty_item_pairs in user_to_difficulty_item_pairs.items():
    prev_timestamp = 0
    prev_difficulty = 'nothing'
    for timestamp,difficulty in difficulty_item_pairs:
      if timestamp > prev_timestamp + interval: # time for new experience sample
        prev_timestamp = timestamp
        prev_difficulty = difficulty
        num_correct += 1
        continue
      if prev_difficulty == difficulty:
        num_correct += 1
      else:
        num_incorrect += 1
  percent_correct = num_correct / (num_correct + num_incorrect)
  minutes_to_percent_correct[minutes] = percent_correct
#print(num_correct)
#print(num_incorrect)
#print(num_correct / (num_correct + num_incorrect))
```

```python
json.dump(minutes_to_percent_correct, open('minutes_to_percent_correct.json', 'wt'))
```

```python
hour_fractions_to_percent_correct = []
for minutes,percent_correct in enumerate(minutes_to_percent_correct):
  hours = minutes / 60
  percent_correct = 100 * percent_correct
  hour_fractions_to_percent_correct.append((hours, percent_correct))
```
