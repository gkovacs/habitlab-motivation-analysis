```python
uninstall_feedback = get_collection_items('uninstall_feedback')
```

```python
print(len(uninstall_feedback))
```

```python
reason_counts = Counter()
num_users = 0
for x in uninstall_feedback:
  if 'uninstall_reasons[]' not in x:
    continue
  num_users += 1
  reasons = x['uninstall_reasons[]']
  if type(reasons) == str:
    reasons = [reasons]
  for reason in reasons:
    if reason == '':
      continue
    reason_counts[reason] += 1
  #break
```

```python
feedback_set = set()
idx = 0
for x in uninstall_feedback:
  if 'feedback' not in x:
    continue
  feedback = x['feedback']
  if 'geza' in feedback:
    continue
  if 'bernstein' in feedback:
    continue
  if feedback == '':
    continue
  #if 'Click here to finish setting' not in feedback:
  #  continue
  if feedback in feedback_set:
    continue
  feedback_set.add(feedback)
  #print('\\textit{\\begin{verbatim}' + feedback + '\end{verbatim}}')
  idx += 1
  print(str(idx) + ': ' +  feedback)
  print('')
  #break
```

```python
print(num_users)
print(sum(reason_counts.values()))
```
