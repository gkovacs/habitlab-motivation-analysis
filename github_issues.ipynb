````python
from github import Github
````

````python
g = Github()
````

````python
repo = g.get_repo("habitlab/habitlab")
open_issues = repo.get_issues()
````

````python
for issue in open_issues:
  title = issue.title
  number = issue.number
  body = issue.body
  if '[User Feedback] ' not in title:
    continue
  title = title.replace('[User Feedback] ', '')
  if 'A user submitted the following via HabitLab\'s built-in Feedback form:' not in body:
    continue
  body = body.replace('A user submitted the following via HabitLab\'s built-in Feedback form:', '')
  if '<img' in body:
    body = body[:body.index('<img')]
  if '```' in body:
    body = body[:body.index('```')]
  body = body.strip()
  if body == '':
    continue
  print('https://github.com/habitlab/habitlab/issues/' + str(number))
  #print(title)
  print(body)
  print()
````
